Q ) N 개의 랜덤 숫자가 주어졌을 때, 해당 숫자들로 만들 수 있는 증가 수열의 최대 길이를 계산하시오. 

A ) (1) N 개 까지의 숫자들을 차례대로 체크한다. 

**(2) 리스트에서 특정 원소 a 보다 앞 쪽에 있는 원소들 중 a 보다 작은 값을 가지는 원소가 있다면, 해당 원소들 중 가장 큰 증가 수열 값에 1 을 더한 값이 원소 a 가 만들 수 있는 최대 증가 수열 값이다.**

```python
n = int(input())
```

```python
arr = list(map(int, input().split()))
```

```python
chk = [0] * len(arr)
```

```python
chk[0] = 1
res = 0
```

```python
for i in range(1, n):
  max = 0
  for j in range(i-1, -1, -1) :
    if arr[i] > arr[j] and max < chk[j] :
      max = chk[j]
  chk[i] = max + 1
  if res < chk[i] : res = chk[i]
```

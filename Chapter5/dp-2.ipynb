## DP - 2

Q ) N 개의 계단을 한번에 1개 또는 2개만 오른다고 할때, 계단을 오를 수 있는 방법의 가지 수를 구하시오. 

A ) 점화식 f(n) = f(n-1) + f(n-2) 를 DFS 와 메모이제이션을 이용해 계산한다. 


```python
n = int(input())
```

```python
chk = [0] * (n+1)
```

```python
chk[1] = 1
chk[2] = 2
```

```python
def dfs(x) :
  if chk[x] > 0 :
    return chk[x]
  if x == 1 or x == 2:
    return x
  else :
    dfs(x-1)
    chk[x] = chk[x-1] + chk[x-2]
    return chk[x]
```

```python
dfs(n)
```

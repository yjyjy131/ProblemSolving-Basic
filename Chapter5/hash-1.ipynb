***Q )*** 특정 문자열의 존재 여부 찾기 

***A )***
Dictionary 로 구현할 것

```python
from collections import deque
```

```python
inputs = input()
```

```python
input_list = inputs.split()
```

```python
input_list
```

```python
n = int(input_list[0])
memo = input_list[1:n+1]
article = input_list[n+1:]
```

```python
m_dict = { k:v for v, k in enumerate(memo) }
```

```python
for k, v in m_dict.items():
  if k not in article : print(k)
```

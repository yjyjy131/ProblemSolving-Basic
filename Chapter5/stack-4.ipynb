***A )***

- 어떤 조건에서 stack 에 pop 하고 append 를 해줄지 잘 구분할 것

1. 숫자인지 문자인지 구분
   - 숫자라면 바로 stack 에 넣는다. 
   - 문자라면 stack 의 두 수를 꺼내 연산하고, 연산한 값을 다시 넣는다. 
     
     **( 이때, 후위연산식이므로 stack 에서 나중에 꺼낸 문자가 앞에 와야함. )**
   - 모든 input 문자가 소진되면 반복문을 끝낸다. 

```python
inputs = input()
```

```python
input_list = list(inputs)
```

```python
stack = []
```

```python
for v in input_list:
  if v.isdecimal():
    stack.append(int(v))
  else :
      num_a = int(stack.pop())
      num_b = int(stack.pop())
      if v == '+': stack.append(num_b+num_a)
      elif v == '-': stack.append(num_b-num_a)
      elif v == '*': stack.append(num_b*num_a)
      elif v == '/': stack.append(num_b/num_a)
```

Q ) N*N 의 맵이 있을 때, 각 맵에 이동시 소요되는 에너지가 표기된다. 최소 에너지를 사용해 최단 거리로 이동해야할 때, 소요되는 최소량의 에너지를 계산하시오. ( 단, 이동시 오른쪽 또는 아래쪽의 두 방향으로만 움직인다. )

A ) (1) 보드를 이동하면서 해당 위치까지 오는데 필요한 최소 에너지 소비량을 dp 에 기록한다.  

(2) dp 의 맨 윗줄, 맨 왼쪽줄은 이동 가능한 방향이 한방향뿐이므로 이전 dp 값에서 현재 보드값의 에너지를 더한 값이 최소값이다.

(3) 나머지 dp 값들은 위쪽 또는 왼쪽 방향 중 더 작은 값의 방향에서 dp 값을 만들어낸다. 

```python
n = int(input())
```

```python
board = [ list(map(int, input().split()))  for _ in range(n)]
```

```python
dp = [ [0] * n for _ in range(n) ] 
```

```python
dp[0][0] = board[0][0]
```

```python
for i in range(1, n) : 
  dp[0][i] = dp[0][i-1] + board[0][i]
```

```python
for i in range(1, n) :
  dp[i][0] = dp[i-1][0] + board[i][0]
```

```python
for i in range(1, n) :
  for j in range(1, n) :
    if dp[i-1][j] + board[i][j] < dp[i][j-1] + board[i][j] :
      dp[i][j] =  dp[i-1][j] + board[i][j]
    else :
      dp[i][j] = dp[i][j-1] + board[i][j]
```

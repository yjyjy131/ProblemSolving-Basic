## Floyd-Warshall Algorithm - 1 ✨

Q ) 노드 수 M 와 각 노드를 연결하는 라인 수 N 가 주어지고, 각 노드를 연결하는 라인들의 가중치 정보가 N 개 주어질 때, 특정 노드에서 특정 노드로 갈 수 있는 최단 경로의 가중치를 출력하시오. ( 단, 갈 수 있는 경로가 없는 경우에는 M 으로 표기 )


A ) M 을 처음부터 기록하지 않고, 아주 큰 수 5000 으로 표기해야 한다. 

그래야 min 값을 구하면서 비교하고, 값을 채워넣는 과정이 수월하기 때문이다. 

**또한, 플로이드 워샬에서 [1][3] 의 값은 1-3 1-2-3 1-3-2 등 1에서 3으로 갈 수 있는 모든 경로에서 최단 가중치 값을 의미한다는 점을 명심해야 한다.** 

```python
node, line = map(int, input().split(가)
```

```python
line_value = [ [5000] * node  for i in range(node) ]
```

```python
for l in range(line) : 
  i, j, v = map(int, input().split())
  line_value[i-1][j-1] = v
```

```python
for i in range(node) : line_value[i][i] = 0
```

```python
for i in range(node) :
  for j in range(node) :
    for k in range(node) :
      line_value[i][j] = min(line_value[i][j], line_value[i][k] + line_value[k][j])
```

```python
for i in range(node) :
  for j in range(node) :
    if line_value[i][j] == 5000 :
      print('M', end = ' ')
    else :
      print(line_value[i][j], end=' ')
  print()
```

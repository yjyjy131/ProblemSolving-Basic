## Knapsack - 1 ✨

Q ) 보석의 종류 N 과 가방의 무게제한 W, 그리고 N개 보석들의 각 무게들과 가중치 값들이 주어질 때 가방에 담을 수 있는 보석들의 최대 가중치 값을 계산하시오. 

A ) (1) dp 를 체크하되, dp[j] 에서 j 는 가방의 무게 제한이다.  

(2) dp[j-무게] 에는 이전에 기록된 j-무게 가방에 담을 수 있는 보석의 최대 가중치가 기록되어 있다. DP 방식을 사용해 dp[j-무게] + 보석 가중치 값과 현재 dp[j] 값을 비교하여 더 큰 보석 가중치 값을 기록한다.

```python
n, w = map(int, input().split())
```

```python
jwe = [ list(map(int, input().split())) for _ in range(n) ]
```

```python
dp = [0] * (w+1)
```

```python
dp
```

```python
for i in range(n) :
  jwe_w = jwe[i][0]
  jwe_v = jwe[i][1]
  for j in range(w+1) :
    if j - jwe_w >= 0 :
      res = dp[j-jwe_w] + jwe_v
      if dp[j] < res : dp[j] = res
```

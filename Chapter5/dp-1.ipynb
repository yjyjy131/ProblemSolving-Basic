## DP - 1 ✨

Q ) Nm 의 리본을 1m, 2m 로만 리본을 자르려 할 때 가능한 방법의 수는 ?


### 풀이 1. Bottom Up

A ) 1m, 2m 의 리본들로  1m, 2m, ,,, Nm 의 리본을 만들 때 사용 가능한 방법의 수를 구하면 점화식이 도출된다. 

(1) 1m, 2m 의 리본으로 1m 의 리본을 만들 수 있는 가지수 (1m) 1가지

(2) 1m, 2m 의 리본으로 2m 의 리본을 만들 수 있는 가지수 (1m+1m) (2m) 2가지

(3) 1m, 2m 의 리본으로 3m 의 리본을 만들 수 있는 가지수 (1m+1m+1m) (2m+1m) (1m+1m) 3가지

(4) 1m, 2m 의 리본으로 4m 의 리본을 만들 수 있는 가지수 (1m+1m+1m+1m) (2m+2m) (1m+2m+2m) (2m+1m+2m) (2m+2m+1m) 5가지

위와 같은 방식으로 점화식이 도출된다. 
f(n) = f(n-1) + f(n-2)

```python
n = int(input())
```

```python
ribbon = [0] * n
```

```python
ribbon[0] = 1
ribbon[1] = 2
```

```python
for i in range(n-2) :
  ribbon[i+2] = ribbon[i] + ribbon[i+1] 
```

### 풀이 2. Top Down

A ) DFS 의 방식으로 계산하되, 계산 결과값을 리스트에 기록하고 참조하는 메모이제이션 방식을 사용한다. 

Screenshot_20210722-174846_Samsung Notes.jpg

```python
n = int(input())
```

```python
chk = [0] * (n+1)
```

```python
chk[1] = 1
chk[2] = 2
```

```python
def dfs(x) :
  if chk[x] > 0 :
    return chk[x]
  if x > 2 :
    dfs(x-1)
    chk[x] = chk[x-1] + chk[x-2]
    return chk[x]
  else :
    return x
```

```python
dfs(n)
```

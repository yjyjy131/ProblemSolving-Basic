### Recursion



***Q )*** 십진수를 입력하면 이진수를 출력하는 재귀 함수 작성

***A )*** 재귀함수 작성시 리스트에 append 하는 방식 사용 x 결과만 바로 출력하는 것이 훨씬 빠름

```python
inputs = int(input())
```

```python
results = []
```

```python
def recursion(x):
  if x == 0 or x == 1:
    print(x%2, end='')
    # results.append(x%2)
    return 
  else :
    b = x%2
    x = x//2
    recursion(x)
    print(b, end='')
    # results.append(x%2)
```

```python
recursion(inputs)
```

```python
''.join(list(map(str,results)))
```

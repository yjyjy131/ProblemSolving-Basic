### Heap-2



***Q )*** 최대힙 ( 0 입력시 최소값 출력, -1 입력시 종료 )

***A )*** 내장모듈 heapq 사용

- heap 의 대상이 되는 리스트 필요

- heapq.heappush(리스트, (-값, 값 )) : 파이썬의 내장모듈 힙은 최소힙으로 구현되어있기 때문에 최대힙으로 변형하여 사용하기 위해 (-값, 값) 튜플 값으로 최소힙 구조를 활용한다. 

- heapq.heappop(리스트)[1] : 튜플 값에서 원래 값만을 가져와야한다. 

```python
import heapq
```

```python
inputs = input()
```

```python
input_list = list(map(int, inputs.split()))
```

```python
input_list
```

```python
hp = []
```

```python
for v in input_list:
  if v == -1 :
    break
  elif v == 0 :
    if len(hp) == 0:
      print(-1)
    else :
      print(heapq.heappop(hp)[1])
  else :
    heapq.heappush(hp, (-v, v))
```

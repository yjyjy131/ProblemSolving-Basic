Q ) N 개의 돌다리를 한번에 1개 또는 2개만 밟는다고 할 때, 돌다리를 모두 건너 땅으로 갈 수 있는 방법의 수를 구하시오 

A ) 점화식 f(n) = f(n-1) + f(n-2) 를 와 메모이제이션 사용

```python
n = int(input())
```

```python
chk = [0] * (n+1)
```

```python
chk[0] = 1
chk[1] = 2
```

```python
for i in range(2, n+1) :
  chk[i] = chk[i-1] + chk[i-2]
```

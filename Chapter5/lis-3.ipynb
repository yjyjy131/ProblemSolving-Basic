Q ) BOJ 2655 가장 높은 탑 쌓기  https://www.acmicpc.net/problem/2655

A ) (1) 벽돌의 번호를 출력해야 하기 때문에 입력을 받을 때 벽돌의 번호도 함께 기록한다. 

(2) 벽돌의 밑면 또는 넓이 중 하나를 기준으로 소팅한다. 

(3) 하나의 기준으로는 이미 소팅되었으므로, 나머지 하나의 기준으로 대소를 비교한다. 동시에 높이를 더하면서 모든 원소들의 높이 dp 를 기록한다. ( 대소 조건을 만족할때만 dp 계산 )

**(4) 벽돌 번호를 찾기 위해 높이 dp 리스트를 높은 순에서 낮은 순으로 탐색한다. 특정 높이 a 가 max 값이라고 할 때, 해당 dp 인덱스와 같은 인덱스를 가지는 bricks 의 원소가 사용되었다고 볼 수 있다. 해당 벽돌의 높이를 a 에서 빼고 위와 같은 방식으로 벽돌 번호들을 탐색한다.**

```python
n = int(input())
```

```python
bricks = [(0, 0, 0, 0)]
```

```python
for i in range(1, n+1) :
  a, h, w = map(int, input().split())
  bricks.append((i, a, h, w))
```

```python
bricks.sort(key = lambda x:x[3])
```

```python
bricks
```

```python
dp = [0] * (n+1)
```

```python
for i in range(1, n+1) :
  for j in range(0, i) :
    if bricks[i][1] > bricks[j][1] :
      dp[i] = max(dp[i], dp[j] + bricks[i][2])
```

```python
dp
```

```python
max_v = max(dp)
idx = n 
results = []
```

```python
while idx > 0 :
  if max_v == dp[idx] :
    results.append(bricks[idx][0])
    max_v -= bricks[idx][2]
  idx -= 1
```

```python
results.reverse()
```

```python
print(len(results))
```

```python
for v in results : print(v)
```

### Queue-3



***Q )*** 특정 문자열 안에서 문자가 CAB 순서대로 나타나는지 체크하시오. 

***A )***
Deque 로 구현할 것

```python
from collections import deque
```

```python
inputs = input()
```

```python
input_list = inputs.split()
```

```python
input_list
```

```python
pivot = input_list[0]
n = int(input_list[1])
target = input_list[2:]
```

```python
for t in target :
  dq = deque(t)
  pivot_d = deque(pivot)

  for v in dq:
    if v == pivot_d[0] :
      pivot_d.popleft()
      if len(pivot_d) == 0:
        print('YES')
        break

  if len(pivot_d) != 0:
    print('NO')
```

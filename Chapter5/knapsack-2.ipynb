## Knapsack - 2 

Q ) BOJ 2294 - 동전 2 https://www.acmicpc.net/problem/2294

A ) (1) dp 를 체크하되, dp[j] 에서 j 는 만들 수 있는 금액의 합이다.  

(2) dp[j-현재동전값] 에는 이전에 기록된 j-현재동전값 을 만들 수 있는 최소의 동전 개수가 기록되어 있다. 따라서 dp[j] 에는 현재 사용되는 동전 1개와, dp[j-현재동전값] 의 최소 동전 값을 더한 값이 기록된다. 

(3) 만약 dp 의 마지막 값이 dp 를 정의할 때의 초기값과 같은 값이라면, 현재 동전 구성으로 동전합을 만들 수 없는 것이므로 -1 (불가능) 을 출력한다. 

```python
n, sum_c = map(int, input().split())
```

```python
coins = []
```

```python
for _ in range(n) : coins.append(int(input()))
```

```python
dp = [100] * (sum_c+1)
```

```python
for c in coins : 
    if c < len(dp) : dp[c] = 1
```

```python
for c in coins :
  for i in range(1, sum_c+1):
    if i-c > 0 and i-c < len(dp) :
        tmp_v = 1 + dp[i-c]
        if dp[i] > tmp_v : dp[i] = tmp_v  
```

```python
if dp[-1] == 100 :
  print(-1)
else :
  print(dp[-1])
```

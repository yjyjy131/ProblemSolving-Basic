## Knapsack - 4 

Q ) 문제의 개수 n 개와 제한시간 t 가 주어졌을때, 각 문제의 점수와 풀이 시간이 n 개 주어진다. 이때, 최대로 얻을 수 있는 문제의 점수를 계산하시오. 

A ) (1) dp 를 체크하되, dp[j] 에서 j 는 문제를 풀이하는데 사용 가능한 시간이다.  

(2) dp[j-문제풀이시간] 에는 이전에 기록된 j-문제풀이시간 안에 얻을 수 있는 최대 값의 점수가  기록되어 있다. 따라서 dp[j] 에는 현재 문제를 풀이했을 때의 점수와, dp[j-문제풀이시간] 의 최대 점수 값을 더한 값이 기록된다.

```python
n, t = map(int, input().split())
```

```python
s = []
```

```python
for _ in range(n) : s.append(list(map(int, input().split())))
```

```python
dp = [0] * (t+1)
```

```python
for i in range(n) :
  curr_s = s[i][0]
  curr_t = s[i][1]
  for j in range(1, t+1) :
    if j - curr_t >= 0 :
      tmp_s = dp[j-curr_t] + curr_s
      if dp[j] < tmp_s :
        dp[j] = tmp_s
```

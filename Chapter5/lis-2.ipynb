## LIS - 2 

Q ) BOJ 11722 가장 긴 감소하는 부분 https://www.acmicpc.net/problem/11722

N 개의 랜덤 숫자가 주어졌을 때, 해당 숫자들로 만들 수 있는 감소 수열의 최대 길이를 계산하시오. 

A ) 증가 수열과 동일한 방식으로 풀이한다. 

```python
n = int(input())
```

```python
arr = list(map(int, input().split()))
```

```python
arr
```

```python
chk = [0] * n
```

```python
chk[n-1] = 1
res = 1
```

```python
for i in range(n-1, -1, -1) :
  max = 0
  for j in range(n-1, i, -1):
    if arr[j] < arr[i] and chk[j] > max :
      max = chk[j]
  chk[i] = max + 1
  if res < chk[i] : 
    res = chk[i]
```

## Knapsack - 3 

Q ) BOJ 11047 - 동전 0 https://www.acmicpc.net/problem/11047

A )  dp 에 기록하는 방법으로 풀이할 때, 메모리 초과가 발생한다면 다음과 같은 방법으로 dp 값을 기록하지 않고 풀이할 수 있다. 

(1) 동전의 값이 큰 동전부터 작은 것으로 차례대로 체크한다. 

(2) 목표 금액 k 가 0 이 아니라면 다음의 과정을 거친다. 

(3) 목표 금액 k 를 현재 동전값으로 나누어 사용가능한 최소의 동전 수를 num 에 더한다. 

(4) 만들어야 하는 남은 목표 금액 k 를 나머지 연산을 통해 계산해준다. 

```python
n, k = map(int, input().split())
```

```python
c  = []
```

```python
for _ in range(n) : c.append(int(input()))
```

```python
num = 0
```

```python
for i in range(n-1, -1, -1):
  if k == 0:
    break

  num += k // c[i]
  k %= c[i]
```

## Floyd-Warshall Algorithm - 1 ✨

Q ) 노드 수 M 와 각 노드를 연결하는 라인 수 N 가 주어지고, 각 노드를 연결하는 라인들의 가중치 정보가 N 개 주어질 때, 특정 노드에서 특정 노드로 갈 수 있는 최단 경로의 가중치를 출력하시오. ( 단, 갈 수 있는 경로가 없는 경우에는 M 으로 표기 )


A ) M 을 처음부터 기록하지 않고, 아주 큰 수 5000 으로 표기해야 한다. 

그래야 min 값을 구하면서 비교하고, 값을 채워넣는 과정이 수월하기 때문이다. 

**또한, 플로이드 워샬에서 [1][3] 의 값은 1-3 1-2-3 1-3-2 등 1에서 3으로 갈 수 있는 모든 경로에서 최단 가중치 값을 의미한다는 점을 명심해야 한다.** 

```python
num = int(input())
```

```python
all_v = [[5000] * num for i in range(num)] 
```

```python
for i in range(num) : all_v[i][i] = 0
```

```python
while 1 :
  i, j = map(int, input().split())
  if i != -1 and j != -1 :
    all_v[i-1][j-1] = 1
    all_v[j-1][i-1] = 1
  else :
    break
```

```python
for i in range(num) :
  for j in range(num) :
    for k in range(num) :
      all_v[i][j] = min(all_v[i][j], all_v[i][k]+all_v[k][j])
```

```python
results = []
```

```python
for i in range(num):
  results.append(max(all_v[i]))
```

```python
res = min(results)
```

```python
print(f'{res} {results.count(res)}')
```

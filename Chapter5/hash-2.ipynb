### Hash-2



***Q )*** 아나그램 / 특정 두 문자열에 포함된 문자들의 종류와 갯수가 같은지 체크

***A )*** Dictionary 로 구현

딕셔너리[키] : 값 반환 or 해당 키 값에 값 삽입 가능

**딕셔너리.get(키, 0 ) : 딕셔너리 안에 키가 있으면 해당하는 값을, 없으면 0을 반환한다.**



```python
from collections import deque
```

## 1번 풀이

```python
inputs = input()
```

```python
input_list = inputs.split()
```

```python
input_list
```

```python
p = { k:input_list[0].count(k) for k in input_list[0] }
```

```python
c = { k:input_list[1].count(k) for k in input_list[1]}
```

```python
if p == c : print('True') 
else : print('False')
```

## 2번 풀이

```python
v_dic = dict()
```

```python
for k in input_list[0]:
  v_dic[k] = v_dic.get(k, 0) + 1
```

```python
for k in input_list[1]:
  v_dic[k] = v_dic.get(k, 0) - 1
```

```python
for k in v_dic :
  if v_dic.get(k) > 0 :
    print('False')
    break

else :
  print('True')
```

## DFS-20
Q ) N*N 의 등산로 지도가 있을 때, 가장 낮은 지점에서 가장 높은 지점으로 갈 수 있는 등산 경로의 개수를 구하시오. ( 단, 경로를 만들 때는 현재 위치보다 낮은 위치로 갈 수 없으며 계속해서 올라가야 한다. )

A) DFS 를 활용해 상하좌우 각 위치의 높이를 비교해가며 문제를 해결

```python
from collections import deque
```

```python
n = int(input())
```

```python
box = []
```

```python
for i in range(n): box.append(list(map(int, input().split())))
```

```python
chk = [[0] * n] * n
```

```python
chk[0][0] = 1
```

```python
x_s = [-1, 0, 1, 0]
y_s = [0, 1, 0, -1]
```

```python
min_v = min(map(min, box))
```

```python
strt = [0, 0]
```

```python
for i, row in enumerate(box) :
  if min_v in row :
    strt[0] = i
    strt[1] = row.index(min_v)
```

```python
max_v = max(map(max, box))
```

```python
fin = [0, 0]
```

```python
for i, row in enumerate(box) :
  if max_v in row:
    fin[0] = i
    fin[1] = row.index(max_v)
```

```python
cnt = 0
```

```python
def dfs(curr) :
  if curr == fin :
    global cnt 
    cnt += 1
  else :
    for i in range(4) :
      x = curr[0] + x_s[i]
      y = curr[1] + y_s[i]

      if 0 <= x < n and 0 <= y < n  and  box[x][y] > box[curr[0]][curr[1]] :
        
        dfs([x, y])
        
```

```python
dfs(strt)
```

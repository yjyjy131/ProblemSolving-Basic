## BFS-4 ✨
Q ) BOJ 7576번 토마토가 익는 날 https://www.acmicpc.net/problem/7576

A) BFS 를 사용해야 한다.  **DFS 를 사용한다면, 박스 안에 익어있는 토마토가 2개 이상일 경우를 동시에 세어나갈 수 없기 때문이다.**

```python
from collections import deque
```

```python
n, m = map(int, input().split())
```

```python
2 2
1 -1
-1 1
```

```python
box = [ list(map(int, input().split())) for _ in range(m)]
```

```python
box
```

```python
x_v = [-1, 0, 1, 0]
y_v = [0, 1, 0, -1]
```

```python
dq = deque()
```

```python
for i in range(m):
  for j in range(n):
    if box[i][j] == 1:
      dq.append([i, j])
```

```python
dq
```

```python
dis = [[0]*n for _ in range(m)]
```

```python
dis
```

```python
while dq :
  now = dq.popleft()
  for i in range(4) :
    x = now[0] + x_v[i]
    y = now[1] + y_v[i]

    if 0 <= x < m and 0 <= y < n and box[x][y] == 0 :
      dis[x][y] = dis[now[0]][now[1]] + 1
      box[x][y] = 1
      dq.append([x, y])
```

```python
chk = False
```

```python
for row in box:
  if 0 in row : chk = True
```

```python
if chk :
  print(-1)
else :
  print(max(map(max, dis)))
```

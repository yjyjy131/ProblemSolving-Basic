## DFS-23 ✨
Q ) 안전지대 탐색 https://www.acmicpc.net/problem/2468

A) DFS 를 활용해 상하좌우를 돌면서 섬의 안전지대를 체크한다. 

이때, 비의 범위는 섬의 높이 범위가 1부터 100 이므로 0 ( 비가 아예 오지 않는 경우 ) ~ 99 ( 100 은 체크해도 가장 작은 섬의 갯수가 되므로 체크하지 않아도 됨 ) 이 된다.

또한, 모든 지대가 아예 잠기지 않는 경우 ( 즉 답이 0 으로 도출되는 경우 ) 는 모든 지대를 1로 봐야하므로 해당 경우의 예외처리를 해주어야 한다. 

```python
n = int(input())
```

```python
box = []
```

```python
box = [ list(map(int, input().split())) for _ in range(n)]
```

```python
x_v = [-1, 0, 1, 0]
y_v = [0, 1, 0, -1]
```

```python
def dfs(x, y, v) :
  chk[x][y] = 1

  for m in range(4):
    xx = x + x_v[m]
    yy = y + y_v[m]

    if 0 <= xx < n and 0 <= yy < n and box[xx][yy] > v and chk[xx][yy] == 0:
      dfs(xx, yy, v)
```

```python
cnt = 0
result = 0
```

```python
for val in range(100) :
  chk = [[0]*n for _ in range(n)]
  cnt = 0
  for i in range(n):
    for j in range(n):
      if box[i][j] > val and chk[i][j] == 0:
        cnt += 1
        dfs(i, j, val)
  result = max(result, cnt)
  if cnt == 0 :
    break
```

```python
if result == 0 :
  print(1)
else :
  print(result)
```

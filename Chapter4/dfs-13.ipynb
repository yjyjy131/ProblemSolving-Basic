## DFS - 13 ✨

Q ) 문제 수 N 을 제한시간 k 분 안에 풀어야 할때, 최대로 얻을 수 있는 점수를 계산하시오.

A) 문제 1 을 푼다/안푼다 => 문제 2 을 푼다/안푼다 => 문제 3을 푼다/안푼다 => .... 의 순서로 DFS 를 이용해 구현한다. 

```python
inputs = input()
```

```python
input_list = [int(v) for v in inputs.split()]
```

```python
n = input_list[0]
lim_t = input_list[1]
tmp = input_list[2:]
```

```python
s = [ tmp[i*2] for i in range(len(tmp)//2) ]
t = [ tmp[i*2+1] for i in range(len(tmp)//2) ]
```

```python
chk = [0] * n
```

```python
results = []
```

```python
def dfs(x) :
  if x == n :
    time_t = 0
    total_s = 0

    for i, v in enumerate(chk) :
      if v == 1 : time_t += t[i]
    
    if time_t <= lim_t : 
      for i, v in enumerate(chk) :
        if v == 1 : total_s += s[i]

      results.append(total_s)

  else :
    chk[x] = 1 
    dfs(x+1)

    chk[x] = 0
    dfs(x+1)
```

```python
dfs(0)
```

```python
max(results)
```

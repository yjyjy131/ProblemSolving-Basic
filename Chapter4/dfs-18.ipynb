Q ) 알파벳 A에는 1로, B에는 2로 이렇게 해서 Z에는 26을 할당하여 암호 코드를 생성한다. 입력된 코드를 알파벳으로 복원하는 경우의 수와 가지수를 출력하시오.

A) DFS 를 이용하여 알파벳 A~Z 의 26가지를 각 레벨마다 체크하되, 두자리 수인 경우와 한자리 수인 경우의 인덱스를 고려하며 다음 레벨로 넘어가야 한다. 

```python
code = list(map(int, input()))
```

```python
c_len = len(code)
```

```python
code.append(-1)
```

```python
res = [0] * (c_len)
```

```python
cnt = 0
```

```python
def dfs(x, p):
  if x == c_len:
    global cnt
    cnt += 1
    for j in range(p):
      print(chr(res[j]+64), end=" ")
    print()
  else :
    for i in range(1, 27):
      if code[x] == i :
        res[p] = i
        dfs(x+1, p+1)
      elif i >= 10 and code[x] == i//10 and code[x+1] == i%10 :
        res[p] = i
        dfs(x+2, p+1)
```

```python
dfs(0, 0)
```

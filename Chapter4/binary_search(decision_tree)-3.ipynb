### Binary search - 결정 알고리즘 3 

***Q )*** 3개의 케이지에 2마리의 햄스터를 넣는다. 각 케이지의 좌표는 1 5 9 일때, 햄스터들은 최소한 한개 케이지 이상은 거리를 두고 배치해야한다. 가장 가까운 두 햄스터의 거리가 최대가 될 수 있도록 할 때, 거리가 가장 가까운 두 햄스터 케이지 사이의 거리를 구하시오.


***A )***
1. 햄스터간 거리를 이진 탐색을 탐색한다. 

2. while 문에서 min <= max 가 될떄까지 탐색한다. 
   이때, min 은 무조건 1 이고 ( 햄스터를 맨 앞에 한마리 넣는 것이 이득이므로 ), max 는 맨 마지막 원소 값에 해당한다. 

3. min / max 의 가운데 길이 m 을 기준으로 현재 케이지 좌표에서, m 길이 차이로는 몇 마리의 햄스터를 배치할 수 있는 지를 계산한다. 
   - 배치 가능한 햄스터 수 >= 현재 햄스터 수 : 거리를 늘려가며 탐색
   - 배치 가능한 햄스터 수 < 현재 햄스터 수 : 거리를 줄여가며 탐색

```python
inputs = input()
```

```python
input_list = [ int(val) for val in inputs.split()]
```

```python
t_num = input_list[0]
h_num = input_list[1]
input_list = input_list[2:]
```

```python
input_list.sort()
```

```python
lt = 1
rt = input_list[-1]
```

```python
def count_h(mylist, length):
  pivot = mylist[0]
  count = 1

  for val in mylist :
    if val - pivot >= length :
      pivot = val 
      count += 1

  return count
```

```python
while lt <= rt :
  m = ( lt + rt ) // 2

  if count_h(input_list, m) >= h_num:
    result = m
    lt = m + 1
  else :
    rt = m - 1
```

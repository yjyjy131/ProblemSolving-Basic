## BOJ 수 정렬하기 3 - https://www.acmicpc.net/problem/10989

A ) 메모리 제한이 타이트 하기 때문에, 입력 범위의 리스트를 미리 만들어 놓고 입력받은 수의 인덱스를 바로 체크하는 방식을 사용한다. ( 계수 정렬 )

```python
n = int(input())
```

```python
chk = [0] * 10001
```

```python
for i in range(n) :
  chk[int(input())] += 1
```

```python
chk
```

```python
for j in range(1, 10001):
  if chk[j] > 0:
    for _ in range(chk[j]) :
      print(j)
```

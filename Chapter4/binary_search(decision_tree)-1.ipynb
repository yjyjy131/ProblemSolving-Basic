### Binary search - 결정 알고리즘 1 

***Q )*** 길이 123, 412, 113, 583, 298 의 4개의 리본이 있을 때, 이 리본들을 같은 길이로 잘라서 10개의 리본을 만드려 한다. 이때 최대의 리본길이 k 를 구하시오. 

***A )***
1. 리본의 min, max 범위 안에서 binary search 방식으로 리본 길이를 탐색해야 한다.때

2. while 문에서 min max 값이 같아질때까지 탐색한다. 

3. 리본을 자른 후 
   - 만들어진 리본 개수 >= 필요한 개수 : 리본 길이를 늘려가며 탐색
   - 만들어진 리본 개수 < 필요한 개수 : 리본 길이를 줄여가며 탐색


```python
inputs = input()
```

```python
input_list = [ int(val) for val in inputs.split()]
```

```python
has = input_list[0]
need = input_list[1]
input_list = input_list[2:]
```

```python
input_list.sort()
```

```python
max_val = input_list[-1]
min_val = input_list[0]
```

```python
med_val = 0
```

```python
while min_val <= max_val :
  med_val = ( max_val + min_val ) // 2
  tmp = [val // med_val for val in input_list]
  all_count = sum(tmp)

  if all_count >= need :
    result = med_val
    min_val = med_val + 1 
  
  elif all_count < need : 
    max_val = med_val - 1
```

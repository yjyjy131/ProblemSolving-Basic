## DFS - 17 
Q ) N 개의 금액이 다른 동전들을 A, B, C 세 사람에게 배분한다고 할 때, 한 사람이 가지는 최대 금액과 최소 금액의 차가 가장 적을 때의 값을 계산하시오. 단, 세 사람은 모두 다른 금액의 동전을 가지고 있다. 

A) DFS 로 풀이시, 각 노드들을 A, B, C 로 놓고 가지들을 각 동전의 종류로 놓아 동전a을 A/B/C 가 가지는 경우 => 동전b를 A/B/C 가 가지는 경우 => ... 로 나누어 풀이한다 . 

```python
n = int(input())
```

```python
c = []
```

```python
for i in range(n) : c.append(int(input()))
```

```python
c
```

```python
sums = [0] * 3
```

```python
results = 100000
```

```python
def dfs(x):
  if x == n :
    calc = max(sums) - min(sums)
    global results 
    if calc < results :
      tmp = set()
      for v in sums : tmp.add(v)
      if len(tmp) == 3: 
        results = calc
  else :
    for i in range(3):
      sums[i] += c[x]
      dfs(x+1)
      sums[i] -= c[x]
```

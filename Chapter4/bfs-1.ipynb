## BFS - 1 ✨
Q ) 민수는 휴대폰을 잃어버렸다. 민수의 위치 n 과 휴대폰 위치 k 가 주어지고, 민수가 스카이 콩콩을 이용해 한번 이동할 때 +1, -1, +5 만큼씩만 이동한다고 할 때 최소 몇번을 움직여야 휴대폰을 찾을 수 있는지 계산하시오. 

A) BFS 를 사용해 계산한다. 이때, deque 를 사용하여 구현할 수 있다. 

```python
from collections import deque
```

```python
strt, t = map(int, input().split())
```

```python
MAX = 100000
```

```python
dis = [0] * ( MAX + 10 )
chk = [0] * ( MAX + 10 )
```

```python
dis[strt] = 0
chk[strt] = 1
```

```python
dq = deque()
dq.append(strt)
```

```python
while dq :
  now = dq.popleft()
  if now == t :
    print(dis[now])
  else :
    for v in ( now-1, now+1, now+5 ) :
      if 0 < v <= MAX and chk[v] != 1:
        dis[v] += dis[now] + 1
        chk[v] = 1
        dq.append(v)
```

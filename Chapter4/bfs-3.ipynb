## BFS - 3 ✨
Q ) 7*7 사이즈의 미로가 있을 때, 1 은 벽을 0은 길을 의미한다. 최단거리로 도착지점에 가기 위해 필요한 최소한의 거리를 계산하시오.  

A) BFS 를 사용해 계산한다. 이때, 어느 방향으로 나아갈지 deque 를 이용해 구현하고 지나온 방향과 거리수를 따로 체크한다. 

```python
from collections import deque
```

```python
box = []
```

```python
for i in range(7): box.append(list(map(int, input().split())))
```

```python
chk = [[0]*7] * 7
```

```python
dq = deque()
```

```python
dq.append([0, 0])
```

```python
box[0][0] = 1
```

```python
x_s = [-1 , 0, 1, 0]
y_s = [0, 1, 0, -1]
```

```python
while dq :
  pos = dq.popleft()
  
  for i in range(4):
    x = pos[0] + x_s[i] 
    y = pos[1] + y_s[i]

    if 0 <= x <= 6 and 0 <= y <= 6 and box[x][y] == 0:
      box[x][y] = 1
      chk[x][y] = chk[pos[0]][pos[1]] + 1
      dq.append([x, y])
```

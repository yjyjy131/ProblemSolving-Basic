## DFS-25 ✨
Q ) 치킨 배달 https://www.acmicpc.net/problem/15686 

A) 1. 치킨집의 수에 맞추어 치킨집의 조합을 구한다. 

2. 모든 집마다 각 조합의 치킨집까지의 최저 거리 합을 구한다. 

3. 최저 거리 합은 계산할때마다 저장하고, 이들 중에서도 가장 최저가 되는 거리가 정답이 된다. 

```python
n, m = map(int, input().split())
```

```python
5 1
1 2 0 2 1
1 2 0 2 1
1 2 0 2 1
1 2 0 2 1
1 2 0 2 1
```

```python
brd = [ list(map(int, input().split())) for _ in range(n)]
```

```python
pizza = []
house = []
```

```python
for i in range(n):
  for j in range(n):
    if brd[i][j] == 1:
      house.append([i,j])
    elif brd[i][j] == 2:
      pizza.append([i, j])
```

```python
res = [-1] * m
```

```python
chk = [0] * len(pizza)
```

```python
result = 1000000
```

```python
results = []
```

```python
def dfs(x, q):
  if x == m :
    calc = 0

    for h_v in house :
      tmp = []
      for r_i in res :
        tmp.append(abs(pizza[r_i][0]-h_v[0]) + abs(pizza[r_i][1]-h_v[1]))
      calc += min(tmp)
    
    if result > calc : 
      if calc not in results :
        results.append(calc)
    
  else :
    for i in range(q, len(pizza)):
      if chk[i] == 0 :
        chk[i] = 1
        res[x] = i
        dfs(x+1, i)

        chk[i] = 0
```

```python
dfs(0, 0)
```

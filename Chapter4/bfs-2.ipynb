## BFS - 2
Q ) 홀수 N * N 크기의 정사각형 모양의 파인애플 밭이 있다. 다이아몬드 모양의 영역에 있는 파인애플만 수확한다고 할때, 수확가능한 파인애플의 전체 수를 계산하시오. 

예를 들어, 3x3 의 밭이 있고 각 칸마다 [[0,1,2],[3,4,5],[6,7,8]] 개의 파인애플이 있다면 1개 + 3개 + 4개 + 5개 + 7개의 파인애플을 수확한다. 

A) BFS 를 사용해 계산한다. 이때, 계산해야할 밭의 위치(인덱스)를  deque 를 사용하여 구현할 수 있다. 

```python
from collections import deque
```

```python
n = int(input())
```

```python
box = []
```

```python
for i in range(n):
  box.append(list(map(int, input().split())))
```

```python
chk = deque()
```

```python
chk.append(n//2)
```

```python
result = 0
```

```python
result += box[0][n//2]
```

```python
for i in range(1, n):

  if i <= n//2 :
    chk.appendleft(chk[0]-1)
    chk.append(chk[-1]+1)
  elif i > n//2 :
    chk.popleft()
    chk.pop()
  
  for idx in chk :
    result += box[i][idx]
```

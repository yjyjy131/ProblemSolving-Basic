## DFS-22
Q ) N*N 의 지도가 있을 때, 1은 땅을 의미한다. 상하좌우 대각선으로 연결되어있으면 하나의 섬으로 본다고 할때, 섬의 갯수를 구하시오.

A) DFS 를 활용해 상하좌우 대각선으로 섬의 영역이 있는지 체크한다. 

```python
n = int(input())
```

```python
box = []
```

```python
for i in range(n): box.append(list(map(int, input())))
```

```python
board = box
```

```python
board
```

```python
x_v = [-1, -1, 0, 1, 1, 1, 0, -1]
y_v = [0, 1, 1, 1, 0, -1, -1, -1]
```

```python
def dfs(x, y) :
  board[x][y] = 0

  for i in range(8) :
    xx = x + x_v[i]
    yy = y + y_v[i]

    if 0 <= xx < n and 0 <= yy < n and board[xx][yy] == 1 :
      dfs(xx, yy)
```

```python
cnt = 0
```

```python
for i in range(n) :
  for j in range(n) :
    if board[i][j] == 1 :
      dfs(i, j)
      cnt += 1
```

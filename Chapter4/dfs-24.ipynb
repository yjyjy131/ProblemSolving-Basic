Q ) 사다리 타기를 할 때, 목적지가 2인 곳을 가기 위한 출발점을 구하시오. 

A) DFS 를 사용해 탐색하되, 도착점 2 인 곳을 시작점으로 하여 탐색한다. 

```python
lad = [ list(map(int, input().split())) for _ in range(10)]
```

```python
chk = [[0]*10]*10
```

```python
for i in range(10) : 
  if lad[9][i] == 2 : 
    arv = [9, i]
```

```python
arv
```

```python
def dfs(x, y) :
  chk[x][y] = 1
  if x == 0 :
    print(y)
  else :
    if y+1 < 10 and lad[x][y+1] == 1 and chk[x][y+1] == 0:
      dfs(x, y+1)
    elif y-1 > 0 and lad[x][y-1] == 1 and chk[x][y-1] == 0:
      dfs(x, y-1)
    else:
      dfs(x-1, y)
```

```python
dfs(arv[0], arv[1])
```

## DFS-27 ✨

Q ) 퀵 정렬 구현

A ) 



```python
n = int(input())
```

```python
arr = [ int(input()) for _ in range(n)]
```

```python
def dfs(lt, rt) :
  if lt < rt :
    pivot = arr[rt]
    pos = lt 
    for i in range(lt, rt+1):
      if arr[i] < pivot :
        tmp = arr[i]
        arr[i] = arr[pos]
        arr[pos] = tmp
        pos += 1
    arr[rt] = arr[pos]
    arr[pos] = pivot

    dfs(lt, pos-1)
    dfs(pos+1, rt)
```

```python
dfs(0, n-1)
```

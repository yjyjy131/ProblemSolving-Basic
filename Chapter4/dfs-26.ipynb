```python
n = int(input())
```

```python
arr = [ int(input())  for _ in range(n)]
```

```python
arr
```

```python
def dfs(lt, rt):
  if lt < rt :
    mid = (lt+rt)//2
    dfs(lt, mid)
    dfs(mid+1, rt)

    p1 = lt
    p2 = mid+1
    tmp = []
    while p1 < mid+1 and p2 < rt+1 :
      if arr[p1] < arr[p2] :
        tmp.append(arr[p1])
        p1 += 1
      else :
        tmp.append(arr[p2])
        p2 += 1
  
    if p1 < mid+1 :
      tmp = tmp + arr[p1:mid+1]  
    if p2 < rt+1 :
      tmp = tmp + arr[p2:rt+1]

    for i in range(len(tmp)) :
      arr[lt+i] = tmp[i]

    
```

```python
dfs(0, n-1)
```

```python
for v in arr :
  print(v)
```

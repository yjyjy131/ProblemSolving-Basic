A) DFS 로 풀이 할 수 있으나, 시간 초과 => DP 로 풀이해야 시간안에 풀이 가능

각 동전의 종류들을 레벨마다 두고, 그래프의 가지를 해당 동전의 사용 갯수로 놓고 계산한다. 

```python
t = int(input())
k = int(input())
```

```python
c = []
c_num = []
```

```python
for i in range(k) :
  a, b = map(int, input().split())
  c.append(a)
  c_num.append(b)
```

```python
cnt = 0
```

```python
def dfs(x, v):
  if x == k :
    if v == t :
      global cnt
      cnt += 1
    
  else :
    for i in range(c_num[x]+1):
      if v + c[x]*i <= t:
        dfs(x+1, v+c[x]*i)
```

```python
dfs(0, 0)
```

## DFS-21 ✨
Q ) N*N 의 지도가 있을 때, 1은 집에 해당한다. 이때, 상하좌우로 이어져있는 집들은 같은 단지로 간주한다. 지도를 탐색하면서, 단지의 수와 각 단지에 속하는 집들의 수를 계산하시오. 

A) DFS 를 활용해 상하좌우로 집이 있는지 탐색한다. 

(1) 전체 보드를 돌면서 1 있다면 cnt 를 초기화하고, dfs 함수를 호출한다. 

(2) 집이 있어 dfs 가 호출되었으므로, 해당 집의 개수를 하나 카운트하고 보드에 지나온 자리를 0 으로 표시한다. 

(3) 상하좌우를 돌면서 집이 있는지 체크하고, 집이 있다면 dfs 함수를 호출한다. 

(4) dfs 함수 사용이 모두 완료되면 계산했던 총 cnt 를 results 에 추가한다. ( 단지 하나에 있는 집의 개수 ) 

(5) 나머지 반복문을 돌면서, 1 이 있는지 체크하고 해당 원소에서 다시 다른 단지내의 집 개수를 계산한다. 

```python
n = int(input())
```

```python
box = []
```

```python
for i in range(n): box.append(list(map(int, input())))
```

```python
board = box
```

```python
board
```

```python
x_v = [-1, 0, 1, 0]
y_v = [0, 1, 0, -1]
```

```python
def dfs(x, y) :
  global cnt
  cnt += 1
  board[x][y] = 0

  for i in range(4) :
    xx = x + x_v[i]
    yy = y + y_v[i]

    if 0 <= xx < n and 0 <= yy < n and board[xx][yy] == 1 :
      dfs(xx, yy)
```

```python
results = []
```

```python
for i in range(n) :
  for j in range(n) :
    if board[i][j] == 1 :
      cnt = 0
      dfs(i, j)
      results.append(cnt)
```

## DFS - 15 ✨
Q )양팔저울 https://www.acmicpc.net/problem/17610 

A) DFS 로 문제를 풀이 => 가지를 왼쪽 저울 사용 / 오른쪽 저울 사용 / 사용 x 의 3가지로 분리하여 가지를 뻗어나간다. 

이때, 왼쪽과 오른쪽 가지는 서로 대칭을 이루기 때문에 총합이 음수인 경우는 count 하지 않아도 된다. 

```python
n = int(input())
w = list(map(int, input.split()))
```

```python
S = sum(w)
res = set()
```

```python
def dfs(v, x):
  global res
  if x == n :
    if v > 0 and v <= S : res.add(v)

  else :
    dfs(v+w[x], x+1)
    dfs(v-w[x], x+1)
    dfs(v, x+1)
```

```python
dfs(0, 0)
```

```python
print(S - len(res))
```

## DFS - 14
Q ) 7일 안에 각각 N 일, k 원의 보수를 받을 수 있는 일의 목록이 주어진다. 예를 들어 4, 20 주어지면 4일 동안 해당 일을 맡아야하고 20원의 보수를 받는다. 단, 4일동안 해당 일 이외의 일은 맡을 수 없다고 할 때, 최대로 받을 수 있는 보수액을 계산하시오.

A) DFS 로 문제를 풀이하되, 어떤 일을 맡으면 N 일 동안은 다른 일을 할 수 없다는 것에 주의한다. 

```python
inputs = input()
```

```python
input_list = [int(v) for v in inputs.split()]
```

```python
lim = input_list[0]
tmp = input_list[1:]
```

```python
t = [ tmp[i*2] for i in range(len(tmp)//2) ]
m = [ tmp[i*2+1] for i in range(len(tmp)//2) ]
```

```python
chk = [0] * lim
```

```python
results = []
```

```python
def dfs(x) :
  if x == lim : 
    time_t = 0
    total_m = 0

    for i, v in enumerate(chk) :
      if v == 1 : time_t += t[i]

    if time_t <= lim :
      for i, v in enumerate(chk) :
        if v == 1 : total_m += m[i]     
      
      results.append(total_m)

  else :
    chk[x] = 1 
    dfs(x + t[x])

    chk[x] = 0
    dfs(x+1)
```

```python
dfs(0)
```

```python
max(results)
```

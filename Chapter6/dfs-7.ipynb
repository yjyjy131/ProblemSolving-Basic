## DFS - 7 ✨

Q ) 정수 N, k 가 주어질 때 N 까지의 정수를 이용해 k 개의 수를 가지는 중복없는 순열을 만드시오.

A) 

중복 순열과 비슷하게 풀이하되, 앞에서 이미 사용했던 노드를 따로 저장하는 고자ㅓㅇ

** 이미 앞에서 사용헀던 노드를 체크하되, 체크를 모두 다 사용한 후에는 체크를 제거해주어야 한다.  **



```python
inputs = input()
```

```python
input_list = [ int(v) for v in inputs.split()]
```

```python
n = input_list[0]
k = input_list[1]
```

```python
chk = [0] * n
ans = [0] * k
cnt = 0
```

```python
def dfs(x):
  global cnt
  if x == k:
    print(ans)
    cnt += 1
  else :
    for v in range(n) :
      if chk[v] == 0 :
        chk[v] = 1
        ans[x] = v+1
        dfs(x+1)

        chk[v] = 0
```

```python
dfs(0)
```

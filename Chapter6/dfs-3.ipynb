### DFS - 3 




***Q )*** N 이 주어졌을때, 1 ~ N 까지 수의 부분집합 중에서, 부분집합 A + 부분집합 B = 전제칩합이 되는 경우가 있는지 계산하시오. 


***A)*** 바이너리 트리 활용방안 : 특정 원소를 사용할것인지 안할것인지를 체크



```python
inputs = input()
```

```python
input_list = [ int(v) for v in inputs.split()]
```

```python
n = input_list[0]
sets = input_list[1:]
```

```python
chk = [0] * n
results = []
```

```python
def dfs(x):
  if x == len(sets)-1 :
    sum1 = 0
    sum2 = 0
    for i, v in enumerate(chk):
      if v == 1:
        sum1 += input_list[i]
      else :
        sum2 += input_list[i]

    if sum1 == sum2:
      results.append(True)

  else :
    chk[x] = 1
    dfs(x+1)

    chk[x] = 0
    dfs(x+1)
```

```python
dfs(0)
```

```python
if True in results : print(True)
else : print(False)
```

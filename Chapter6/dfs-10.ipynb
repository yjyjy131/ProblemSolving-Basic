## DFS - 10

Q ) 숫자 N, k 가 주어졌을 때 N개의 숫자들로 구할 수 있는 조합들이 있을 때, 각 조합 원소들의 합이 t 값으로 나눠지는 경우의 수를 구하시오. 

예를 들어 3, 2, 1, 2, 3, 3가 주어지면 조합은 (1, 2), (1, 3), (2, 3) 의 3개이고, 타겟값이 3이라면 경우의 수는 (1,2)의 1이다. 

A) 조합을 구하는 방식을 활용한다. 

```python
inputs = input()
```

```python
input_list = [ int(v) for v in inputs.split()]
```

```python
n = input_list[0]
k = input_list[1]
t = input_list[-1]
nums = input_list[2:-1]
```

```python
chk = [0] * n
ans = [0] * k
cnt = 0
```

```python
def dfs(x, q):
  global cnt
  if x == k :
    if sum(ans) % t == 0 : cnt += 1 
  else :
    for v in range(q, n):
      if chk[v] == 0:
        chk[v] = 1
        ans[x] = nums[v]
        dfs(x+1, v)
        chk[v] = 0
```

```python
dfs(0, 0)
```

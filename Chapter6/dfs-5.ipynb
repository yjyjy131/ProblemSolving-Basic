***Q )*** N, k 가 주어질 때, N 까지의 수를 이용해 k 개의 원소를 가지는 중복순열과 그 수를 구하여라. 


***A)*** 중복순열이므로 각 노드가 또 다시 N 개의 노드를 가지는 트리를 만든다.

```python
inputs = input()
```

```python
input_list = [ int(v) for v in inputs.split()]
```

```python
n = input_list[0]
k = input_list[1]
```

```python
chk = [0] * k
```

```python
cnt = 0
```

```python
def dfs(x):
  if x == k :
    print(chk)
    global cnt 
    cnt += 1
  else :
    for i in range(n):
      chk[x] = i+1
      dfs(x+1)
```

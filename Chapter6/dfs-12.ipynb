## DFS - 12 ✨

Q ) 노드 수 N 과 라인 수 K 가 주어질 때, 특정 노드에서 다른 노드로 가는 방향이 주어진다. 노드 1에서 노드 N 까지 갈 수 있는 경우의 수를 구하시오. ( 단, 한 번 방문했던 노드는 다시 방문할 수 없다. )

A) DFS 를 이용해 구현한다. 단, dfs(x) 의 인자 x 는 노드 값을 이용한다. 

```python
from collections import deque
```

```python
inputs = input()
```

```python
input_list = [int(v) for v in inputs.split()]
```

```python
node_n = input_list[0]
line_n = input_list[1]
graphs = input_list[2:]
```

```python
g = [ (graphs[2*v], graphs[2*v+1]) for v in range(len(graphs)//2)]
```

```python
visit = [0] * node_n
```

```python
cnt = 0
```

```python
def dfs(x):
  if x == node_n :
    global cnt
    cnt += 1
  else :
    if visit[x-1] == 0 :
      for v in g :
        if v[0] == x :
          visit[x-1] = 1
          dfs(v[1])
          visit[x-1] = 0
```

```python
dfs(1)
```

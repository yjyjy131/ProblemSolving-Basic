## DFS - 11

Q ) 노드 수 N 과 라인 수 K 가 주어질 때, 각 노드에서 노드로의 가중치 값이 k 개 주어진다. 해당 가중치 그래프의 인접 행렬을 출력하시오. 


A) 만약 1 2 10 이 주어져 있다면 노드 1 에서 노드 2 로 가는 길이 있고, 해당 길의 가중치가 10이라는 뜻이다. 따라서 인접 행렬은 0번 행과 1번 행이 만나는 [0][1] 위치에 가중치 값 10 을 넣어주면 된다. 

노드 1 => 1, 2, 3 / 노드 2 => 1, 2, 3 / ... 과 같은 방식으로 레벨 1, 2 까지만 계산하면 2차원의 인접 행렬을 나타낼 수 있다. 

```python
from collections import deque
```

```python
inputs = input()
```

```python
input_list = [int(v) for v in inputs.split()]
```

```python
node_n = input_list[0]
line_n = input_list[1]
graphs = input_list[2:]
```

```python
g = [ (graphs[3*v], graphs[3*v+1], graphs[3*v+2]) for v in range(len(graphs)//3)]
```

```python
g
```

```python
all_g = deque(g)
```

```python
def dfs(x, curr, chk):
  if x == 2: 
    if len(all_g) != 0 :
      if all_g[0][0] == curr and all_g[0][1] == chk :
        print(all_g[0][2], end=" ")  
        all_g.popleft()
      else :
        print(0, end=" ")
    else :
      print(0, end=" ")
  else: 
    for i in range(node_n) :
      if chk == 0 : curr = i+1
      dfs(x+1, curr, i+1)
    
    print()
```

```python
dfs(0, 0, 0)
```

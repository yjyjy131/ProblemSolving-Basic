### Greedy - 5

***Q )*** 5 개의 원소를 가진 리스트 1 6 7 3 2 이 있다고 할 때, 양 끝의 수 중 하나를 가져와 오름차순 순열 리스트를 만들고, 만들어진 리스트의 길이를 구하시오. 

***A )***
1. 비교 기준값 pivot 생성

2. lt , rt 로 리스트의 양 끝값을 매번 tmp 에 정답 후보를 튜플 형태로 가져온다. 

3. pivot 의 값보다 lt, rt 가 둘 다 큰 경우 모두 정답 후보가 될 수 있는데, 이런 경우 둘 중 작은 수가 정답이 되어야 하므로 tmp 를 오름차순 정렬해 가장 작은 값, 즉 정답을 가져온다. 

4. 다음 비교를 위해 비교군에 해당 정답을 가져오고 lt, rt 값을 알맞게 옮긴다. 

```python
inputs = input()
```

```python
input_list = [ int(val) for val in inputs.split()]
```

```python
all_n = input_list[0]
k = input_list[1:]
```

```python
pivot = 0
```

```python
lt = 0
rt = all_n - 1
```

```python
result = []
```

```python
tmp = []
```

```python
while lt <= rt :
  if k[lt] > pivot:
    tmp.append((k[lt], 'L'))
  
  if k[rt] > pivot:
    tmp.append((k[rt], 'R'))
  
  tmp.sort()

  if len(tmp) == 0:
    break
  else :
    result.append(tmp[0][1])
    pivot = tmp[0][0]
    if tmp[0][1] == 'L' :
      lt += 1
    else :
      rt -= 1 
  
  tmp.clear()
```

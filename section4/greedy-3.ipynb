***Q )*** 창고 10칸에 짐이 8 3 12 4 6 2 11 2 9 7 개씩 쌓여있다. 가장 많은 짐칸의 짐을 가장 짐이 적은 칸으로 옮기는 방식으로 짐을 5번 옮긴다고 했을때, 가장 많이 쌓인 짐과 적게 쌓인 짐의 갯수 차이를 구하시오. 

***A )***
최대 최소의 짐칸의 인덱스를 찾고 +1, -1 로 짐 수를 계산한다. (X)

1. 모든 짐칸들을 오름차순 정렬한다. 

2. 정렬하게 되면, 가장 짐이 많이 쌓인 짐칸이 맨 뒤로, 가장 짐이 적은 짐칸이 맨 앞으로 오게 되므로 맨 뒤의 짐칸에서 짐을 하나빼고, 맨 앞의 짐칸에서 짐을 하나 더한다. 

3. 다시 짐칸들을 오름차순 정렬하고 이 과정을 반복한다. 



## 비효율

```python
inputs = input()
```

```python
input_list = [ int(val) for val in inputs.split()]
```

```python
all_n = input_list[0]
change = input_list[-1]
input_list = input_list[1:-1]
```

```python
for i in range(0, change):
  maxx = max(input_list)
  max_idx = input_list.index(maxx)
  input_list[max_idx] = maxx - 1

  minn = min(input_list)
  min_idx = input_list.index(minn)
  input_list[min_idx] = minn + 1
```

```python
max(input_list) - min(input_list)
```

## 효율

```python
inputs = input()
```

```python
input_list = [ int(val) for val in inputs.split()]
```

```python
all_n = input_list[0]
change = input_list[-1]
input_list = input_list[1:-1]
```

```python
input_list.sort()
```

```python
for _ in range(0, change):
  input_list[0] += 1
  input_list[all_n-1] -= 1
  input_list.sort()
```

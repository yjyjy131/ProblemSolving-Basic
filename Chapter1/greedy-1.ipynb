### Greedy - 1

***Q )*** 회의의 수가 5개 이고, 각 회의의 시작 시간과 끝 시간이 1 4 / 2 3 / 1 6 / 3 7 / 3 6 이라고 할 때, 하나의 회의실을 사용해 진행할 수 있는 최대의 회의 수를 구하시오.


***A )***
1. 회의를 끝나는 순으로 정렬한다. ( 끝나는 시간이 맞물려야 가장 많은 회의 수를 셀 수 있다. ) 

2. 이전 회의의 끝나는 시간 <= 다음 회의의 시작 시간이라면 카운트를 +1 하고, 해당 회의의 끝나는 시간을 다음 회의 시간 비교를 위해 저장한다. 


```python
inputs = input()
```

```python
input_list = [ int(val) for val in inputs.split()]
```

```python
rooms = input_list[0]
times = input_list[1:]
```

```python
times_list = [ ( times[i*2], times[i*2+1] ) for i in range(len(times)//2)]
```

```python
times_list.sort(key=lambda x : (x[1], x[0]))
```

```python
end_t = 0
count = 0
for start, end in times_list :
  if end_t <= start :
    count += 1
    end_t = end 
```

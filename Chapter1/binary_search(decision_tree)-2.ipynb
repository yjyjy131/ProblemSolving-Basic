### Binary search - 결정 알고리즘 2 

***Q )*** 무게가 1, 2, 3, 5, 2, 7, 8, 3, 4 인 9개의 과일을 4개 바구니에 담을때, 몇 kg 까지 담을 수 있는 바구니를 사용해야 하는지 구하시오.  
( 단, 모두 같은 종류의 바구니를 사용해 과일을 담아야 함 )

***A )***
1. 바구니 kg 수를 최소 ( 가장 무거운 과일 무게 ) ~ 최대 ( 전체 과일 무게 합 ) 사이에서 탐색

2. while 문에서 min <= max 가 될떄까지 탐색한다.

3. 특정 바구니 용량을 탐색하고, 해당 용량에 과일을 나눠 담으면 몇개의 바구니를 사용하는지를 계산한다. 
   - 사용해야하는 바구니 개수 >= 사용해야하는 바구니 개수 : 바구니 용량을 늘려가며 탐색
   - 사용해야하는 바구니 개수 < 사용해야하는 바구니 개수 : 바구니 용량을 줄여가며 탐색

* 헷갈렸던 부분

특정 바구니 용량을 선택했을 경우에 몇개의 바구니로 나눠 담게 되는지의 경우 계산

if 문에서도 연산자 값을 사용할 수 있음을 기억 

```python
inputs = input()
```

```python
input_list = [ int(val) for val in inputs.split()]
```

```python
song_num = input_list[0]
split_num = input_list[1]
input_list = input_list[2:]
```

```python
maxx = max(input_list)
```

```python
lt = max(input_list)
rt = sum(input_list)
```

```python
def calc_splited_num (inputs, length):
  tmp = 0
  count = 1

  for val in inputs :
    if tmp + val > length :
      tmp = val 
      count += 1 

    else :
      tmp += val   

  return count
```

```python
while lt <= rt :
  m = ( lt + rt ) // 2

  if m >= maxx and calc_splited_num(input_list, m) <= split_num:
    result = m
    rt = m - 1
  else :
    lt = m + 1
```

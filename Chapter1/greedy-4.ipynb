### Greedy - 4

***Q )*** 몸무게가 77 45 78 64 52 45 인 6명의 사람이 130 kg 까지 최대 2명을 실을 수 있는 놀이기구에 타려고한다. 놀이기구를 최소 몇 대 사용해야하는지 구하시오. ( 단, 놀이기구의 제한 몸무게 보다 무거운 사람은 없다. )

***A )***
Deque : 큐가 비어있지 않으면, 맨 왼쪽의 원소 값을 반환 후 삭제한다. 

1. 사람 몸무게 순으로 내림차순 정렬한다. 

2. 큐 에서 몸무게가 가장 큰 사람과 가장 작은 사람이 놀이기구에 탈 수 있는지 확인한다. 
   - 둘 모두 탈 수 있다면, count +1 하고 둘을 큐 목록에서 뺀다
   - 둘 모두 탈 수 없다면 몸무게가 가장 큰 한 사람만 탈 수 있으므로 count +1 하고 한명만 목룍에서 뺀다. 


```python
from collections import deque
```

```python
inputs = input()
```

```python
input_list = [ int(val) for val in inputs.split()]
```

```python
all_n = input_list[0]
boat_w = input_list[1]
input_list = input_list[2:]
```

```python
input_list.sort(reverse=True)
```

```python
count = 0
```

```python
dq = deque(input_list)
```

```python
while dq:
  if len(dq) == 1 :
    count += 1
    break

  elif dq[0] + dq[-1] > boat_w:
    dq.pop()
    count+= 1

  else :
    dq.pop()
    dq.popleft()    
    count += 1
```

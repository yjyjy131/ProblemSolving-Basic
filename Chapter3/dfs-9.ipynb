## DFS - 9

Q ) 숫자 N, k 가 주어졌을 때, 조합과 그 수를 구하시오. 

예를 들어 3, 2 가 주어지면 조합은 (1, 2), (1, 3), (2, 3) 의 3개이다. 

A) 

앞 문제들의 중복없는 순열을 구하는 방식을 활용하되, 만약 1이 노드의 시작이라면 2, 3 을 2가 노드의 시작이라면 3 만을 계산하는 과정을 추가한다. 



```python
inputs = input()
```

```python
input_list = [ int(v) for v in inputs.split()]
```

```python
n = input_list[0]
k = input_list[1]
```

```python
chk = [0] * n
ans = [0] * k
cnt = 0
```

```python
def dfs(x, q):
  global cnt
  if x == k :
    cnt += 1
    print(ans)
  else :
    for v in range(q, n):
      if chk[v] == 0:
        chk[v] = 1
        ans[x] = v+1
        dfs(x+1, v)
        chk[v] = 0
```

```python
dfs(0, 0)
```

### DFS - 4 




***Q )*** 5마리의 고양이 무게가 각각 192 48 77 36이고, 케이지가 고양이를 담을 수 있는 최대 무게가 300 이라고 할 때, 제일 많이 고양이를 담을 수 있는 무게를 구하시오. 

***A)*** 바이너리 트리 활용방안 : 특정 원소를 사용할것인지 안할것인지를 체크


```python
inputs = input()
```

```python
input_list = [ int(v) for v in inputs.split()]
```

```python
w = input_list[0]
n = input_list[1]
cats = input_list[2:]
```

```python
chk = [0] * n
results = []
```

```python
def dfs(x):
  if x == n :
    sum = 0
    for i, v in enumerate(chk):
      if v == 1 :
        sum += cats[i]

    if sum < w : results.append(sum)

  else :
    chk[x] = 1
    dfs(x+1)

    chk[x] = 0
    dfs(x+1)
```

```python
dfs(0)
```

```python
max(results)
```

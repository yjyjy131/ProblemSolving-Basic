### DFS - 2 ✨



***Q )*** N 이 주어졌을때, 1 ~ N 까지 수의 부분집합 구하기 

***A)*** 바이너리 트리 활용방안 : 특정 원소를 사용할것인지 안할것인지를 체크

```python
inputs = input()
```

```python
n = int(inputs)
```

```python
ch = [0] * (n)
```

```python
def dfs(x):
  if x == n+1 :
    for i, v in enumerate(ch):
      if v == 1 :
        print(i+1, end=" ")
    
    print()

  else :
    ch[x-1] = 1
    dfs(x+1)

    ch[x-1] = 0
    dfs(x+1)
```

```python
dfs(1)
```

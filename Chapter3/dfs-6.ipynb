## DFS - 6 ✨

Q ) 4가지 종류의 동전 1, 2, 3, 4원이 있을때, 가장 적은 수의 동전을 사용해 10원을 만들기 위해서는 총 몇 개의 동전을 사용해야 하는가?

A) 

1번 노드에서 1 2 3 4 번 노드, 또다시 다음 노드에서 모든 노드를 순회하는 방식으로 구현 ( DFS-5번 문제 방식 ) 

**풀이 1 - 트리의 LEVEL 를 활용 : 현재 트리의 level 이 sum 계산에 사용되어진 노드수와 같으므로 level 값을 활용한다. **

풀이 2 - 사용하고 있는 노드 수를 직접 count 


```python
inputs = input()
```

```python
input_list = [ int(v) for v in inputs.split()]
```

```python
n = input_list[0]
coins = input_list[1:-1]
t = input_list[-1]
```

## 풀이 1 - 트리의 LEVEL 을 이용

```python
results = []
```

```python
def dfs(level, sum):
  if sum > t:
    return
  elif sum == t :
    results.append(level)
  else :
    for v in coins:
      dfs(level+1, sum+v)
```

```python
dfs(0, 0)
```

```python
min(results)
```

## 풀이 2 - LEVEL 을 사용하지 않고 직접 사용 노드수를 count

```python
counts = []
```

```python
def dfs(x, sum, cnt):
  if sum > t :
    return
  elif sum == t :
    counts.append(cnt)
  else :
    for v in coins :
      dfs(v, sum+v, cnt+1)
```

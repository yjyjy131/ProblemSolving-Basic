Q ) 파스칼의 삼각형 / 숫자 N, k 가 주어질 때 파스칼의 삼각형을 이용해 1부터 N 까지 조합의 수를 더했을 때 결과값 k 가 도출된다. 이 때 삼각형 맨 윗단의 숫자 조합을 구하시오. 

A) 

1. 문제에서 만약 3 이 주어졌다면, 삼각형의 1층은 1, 2, 3 => 2층은 1+2, 2+3 => 3층은 1+2+2+3 이다. 즉, 마지막에 결국 계산되는 식은 1x1개, 2x2개, 3x1개인데, 각 숫자에 곱해지는 갯수는 N-1 의 이항계수 값 2C0, 2C1, 2C2 와 같다. 

**1부터 N 까지의 각 숫자들은 더해야하는 갯수가 정해져있는데, 그 수는 이항계수와 같음을 이용해야 한다.**

2. N 이 주어지면, N-1 에 대한 콤비네이션 값을 계산해 저장한다. ( 이 값은 바뀌지 않고, 계속해서 사용되기 때문에 )

3. 1부터 N 까지의 수로 만들 수 있는 순열을 모두 생성한다. 

ex ) 123 132 213 231 312 321

4. 생성된 순열과 콤비네이션 값을 더하여 문제에서 요구하는 목표값과 같은지 체크한다. 

### 1. 재귀를 이용한 풀이

```python
inputs = input()
```

```python
input_list = [ int(v) for v in inputs.split()]
```

```python
n = input_list[0]
k = input_list[1]
```

```python
chk = [0] * n
ans = [0] * n
```

```python
combs = [1] * n 
```

```python
def comb(i):
  result= 1
  for _ in range(n-1, n-i-1, -1): result *= _
  return result
```

```python
def factorial(i) :
  result = 1
  for m in range(1, i+1): result *= m
  return result
```

```python
for j in range(1, n-1):
  combs[j] = comb(j) // factorial(j)   
```

```python
exit_chk = False
```

```python
def dfs(x):
  global exit_chk
  if exit_chk == True: return 

  if x == n :
    res = 0
    for i in range(n) :
      res += ans[i] * combs[i]
    if res == k : 
      print(ans)
      exit_chk= True

  else :
    for v in range(n):
      if chk[v] == 0:
        chk[v] = 1
        ans[x] = v+1
        dfs(x+1)

        chk[v] = 0
```

```python
dfs(0)
```

### 2. 라이브러리 이용한 풀이

```python
import itertools as it
```

```python
inputs = input()
```

```python
input_list = [ int(v) for v in inputs.split()]
```

```python
n = input_list[0]
t = input_list[-1]
```

```python
p_list = [ i+1 for i in range(n)]
```

```python
combs = [1] * n
```

```python
def combi(i):
  res = 1
  for k in range(n-1, n-i-1, -1): res *= k
  return res
```

```python
def facto(i):
  res = 1
  for k in range(1, i+1) : res *= k
  return res
```

```python
for i in range(1, n-1):
  combs[i] = combi(i) // facto(i)
```

```python
for permu in it.permutations(p_list):
  res = 0

  for j in range(n):
    res += permu[j] * combs[j]

  if res == t :
    print(permu)
    break
```

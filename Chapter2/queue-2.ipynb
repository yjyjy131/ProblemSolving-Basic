### Queue-2



***Q )*** N 개의 수가 있을때, k 번째 수의 자리를 모니터링한다. 큐를 돌면서 첫번째 수보다 큰 수가 있으면 해당 수를 큐의 마지막으로 넣고, 첫번째 수가 가장 크다면 해당 수는 큐에서 제외한다. 이 과정을 반복할 때 처음 k 번째 수는 몇 번의 연산을 거쳐야 큐에서 제외되는지 구하시오.

***A )***
Deque 로 구현할 것

```python
from collections import deque
```

```python
inputs = input()
```

```python
input_list = list(map(int, inputs.split()))
```

```python
n = input_list[0]
curr = input_list[1]
```

```python
dq = deque(input_list[2:])
```

```python
cnt = 0
```

```python
while dq:
  if dq[0] < max(dq) :
    dq.append(dq.popleft())
    if curr != 0 :
      curr -= 1
    else :
      curr = len(dq)-1
  else :
    dq.popleft()
    cnt += 1
    if curr == 0:
      break
    else :
      curr -= 1
```

### Queue-1



***Q )*** 100 개 의 수가 있고, 1 부터 수를 목록의 맨 뒤로 보낸다. 이때, 4번째 마다 나오는 숫자는 목록에서 제거한다. 마지막 하나의 수만 남을때까지 이 과정을 반복할때 남아있는 숫자를 구하시오. 

***A )***
Deque 로 구현할 것

```python
from collections import deque
```

```python
inputs = input()
```

```python
input_list = list(map(int, inputs.split()))
```

```python
n = input_list[0]
k = input_list[1]
```

```python
dq = deque([ i for i in range(n) ])
```

```python
cnt = 1
```

```python
while len(dq) > 1:
  if cnt == k :
    dq.popleft()
    cnt = 1

  else :
    dq.append(dq.popleft())
    cnt += 1
  
```

### Stack - 1 ✨



***Q )*** 어떤 랜덤수에 대해 N 개 이하의 수를 삭제하여 가장 큰 값의 수를 만드시오.

***A )***
1. 입력 받은 숫자들을 하나씩 돌면서, 스택에 추가할지 말지를 결정한다. 

2. 현재 비교군 숫자이 스택의 맨 위 값보다 크다면, 스택의 맨 위값을 제거한다. 
   수를 제거했다면 제거할때마다 제거 가능 횟수를 체크한다. 

3. 위 과정을 while 문으로 돌면서 스택에서 제거가능한 모든 수를 제거했는지 체크한다. 

4. 현재 비교군 숫자를 스택에 추가한다. ( 제거 대상이라면 다음 비교군이 비교해주어 제거될 것임 )

5. 만약, 제거 횟수가 아직 남아있는 경우에는 남은 횟수만큼 제거되지 못한 리스트에 남은 값들을 스택에 넣어준다.

```python
inputs = input()
```

```python
input_list = inputs.split()
```

```python
rmv = int(input_list.pop())
```

```python
nums =  list(map(int, list(input_list[0])))
```

```python
nums
```

```python
stack = []
```

```python
for n in nums :
  while rmv > 0 and stack and stack[-1] < n :
    stack.pop
    rmv -= 1
  stack.append(n)

if rmv != 0 :
  stack.append(nums[:-n])
```

### Heap-1



***Q )*** 최소힙 ( 0 입력시 최소값 출력, -1 입력시 종료 )

***A )*** 내장모듈 heapq 사용

- heap 의 대상이 되는 리스트 필요

- heapq.heappop(리스트) : 힙의 최소값 반환 = hp[0]

- heapq.heappush(리스트, 값) : 힙에 원소 삽입 

```python
import heapq
```

```python
inputs = input()
```

```python
input_list = list(map(int, inputs.split()))
```

```python
input_list
```

```python
hp = []
```

```python
for v in input_list:
  if v == -1 :
    break
  elif v == 0 :
    if len(hp) == 0:
      print(-1)
    else :
      print(heapq.heappop(hp))
  else :
    heapq.heappush(hp, v)
```

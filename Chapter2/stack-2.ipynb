***Q )*** BOJ 10799 쇠막대기

***A )***
( ( ( ) ( ) ) )

1. ) 앞에 ( 가 있다면, 레이저에 해당

2. ) 앞에 ) 가 있다면, 쇠막대에 해당

3. 괄호를 차례대로 스택에 넣고, () 가 만들어지면 레이저이므로 둘을 pop 한다. 
이 때, 레이저가 잘라서 생성된 막대 수를 세는데, 왼쪽의 결과만 카운트 한다. ( 오른쪽의 결과는 아직 알 수 없으므로 ) 이 수는 스택에 남아있는 괄호의 수와 같다. 

4. 괄호가 ) ) 라면, 레이저가 끝나고 쇠막대의 끝을 알려주는 것이다. 따라서 더 이상 레이저가 막대를 자를 수 없으므로 다 자르고 남은 막대의 수 1 을 더해준다. 

```python
inputs = input()
```

```python
input_list = list(inputs)
```

```python
stack = []
```

```python
cnt = 0
```

```python
for i in range(len(input_list)):
  if input_list[i] == '(':
    stack.append(i)
  else :
    stack.pop()
    if input_list[i-1] == ')':
      cnt += len(stack)
    else :
      cnt += 1
```

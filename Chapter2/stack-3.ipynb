***A )***

- 어떤 조건에서 stack 에 pop 하고 append 를 해줄지 잘 구분할 것

1. 숫자인지 문자인지 구분
   - 숫자라면 바로 stack 에 넣는다. 
   - 문자라면 우선순위별로 다음과 같이 구분한다. 
   1) 괄호 ( 이다. 
     - 괄호 ( 라면 우선순위가 가장 높으므로 stack 에 넣는다. 

   2) 괄호 ) 이다. 
     - ( 를 만날때까지 스택에 쌓인 모든 연산자를 pop 한다. 

   3 ) * / 기호 이다.
     - 스택의 마지막 기호가 * / 라면, 해당 기호가 우선순위가 높으므로 pop
     - 스택의 마지막 기호가 + - 라면 해당 기호는 우선순위가 낮으므로 그대로 둔다
     - 스택의 마지막 기호가 ( 라면 ) 가 나올때까지 기다려야 하므로 그대로 둔다
   
     => 스택의 마지막 기호가 * 또는 / 라면 pop 한다. 
   
     => 현재 기호를 stack 에 넣는다. 

   4 ) + - 기호 이다. 
    - 스택의 마지막 기호가 ( 라면, ) 가 나올때까지 기다려야하므로 stack 에 넣는다. 
    - **스택의 마지막 기호가 + - 라면 pop 해야 한다.** 
    - 스택의 마지막 기호가 * / 라면 pop 해야 한다. 
   
    => 스택의 마지막 기호가 ( 가 아니라면 pop 한다. 
    
    => 현재 기호를 stack 에 넣는다. 

  5 ) 스택에 여전히 남아있는 기호들을 모두 빼준다.

```python
inputs = input()
```

```python
input_list = list(inputs)
```

```python
op = []
results = ""
```

```python
for v in input_list:
  if v.isdecimal():
    results += v
  else :
    if v == '(':
      op.append(v)
    elif v == '*' or v == '/':
      while op and ( op[-1] == '*' or op[-1] == '/'):
        results += op.pop()
      op.append(v)
    elif v == '+' or v == '-':
      while op and op[-1] != '(' :
        results += op.pop()
      op.append(v)
    elif v == ')' :
      while op and op[-1] != '(' :
        results += op.pop()
      op.pop()

while op:
  results += op.pop()  
```
